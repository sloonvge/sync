```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.io as sio
```

```python
def PlotStdMeanCurve(*Xs, size=20):
    xlims = np.linspace(np.log10(0.001), np.log10(1), size)
    colors = ['b', 'r', 'y']
    shapes = ['o--', 'x--', 'v--']
    labels = ['w=0.1', 'w=1', 'w=10']
    i = 0
    for X in Xs:
        XMean = np.mean(X, axis=0)
        XStd  = np.std(X, axis=0)
        
        plt.fill_between([np.power(10, i) for i in xlims], XMean-XStd, XMean+XStd, alpha=0.1, color=colors[i])
        
        plt.semilogx([np.power(10, i) for i in xlims], XMean, shapes[i], color=colors[i], label=labels[i])
        
        i += 1
    
    plt.legend(loc=0)
    plt.show()
```

```python
a = np.random.random((10, 20))
b = np.random.random((10, 20))
c = np.random.random((10, 20))
```

```python
PlotStdMeanCurve(a, b, c)
```
